```python
# initialize setup and calcuations for provided bike dataset, you must press enter to continue
s = setup(data = train, 
          test_data = test, 
          target = 'cnt', 
          fold_strategy= 'timeseries', 
          numeric_features = ['Month', 'Series'], 
          fold = 3, 
          transform_target = True, 
          session_id = 123)
```
